```python
#Linear Regression With One Variable Sample problem of predicting home price in monroe, new jersey (USA)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
df=pd.read_csv("prediction.csv")
df.head()
plt.xlabel("area")
plt.ylabel("prices")
plt.scatter(df.area,df.prices,color="red",marker="+")
#now lookat the distribution we can make use of linear regression
our_model=linear_model.LinearRegression()
our_model.fit(df[['area']],df.prices)
prediction = our_model.predict(np.array([[3500]]))
print(f"Predicted price for an area of 3500: {prediction[0]}")
```
